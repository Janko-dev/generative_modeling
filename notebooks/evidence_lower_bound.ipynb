# Evidence lower bound (ELBO)

```python
import torch
from torch import nn
from torch.nn import functional as F
import matplotlib.pyplot as plt
import numpy as np
```

## Latent Variable Model

The observed data $x$ is a realization from some random variable $X$. We assume the existence of another random variable $Z$, where $X$ and $Z$ are distributed according to a joint distribution $p(X, Z; \theta)$, where $\theta$ parameterizes the distribution. The latent $Z$ remains unobserved, while the data is observed and a realization of $X$. The reason we can assume a latent $Z$ is due to the **law of total probability**:

$$
    p(X; \theta) = \int p(X, Z=z; \theta) dz = \int p(X \;|\; Z=z; \theta) p(Z=z) dz = \mathbb{E}_{z \sim p(Z)}[p(X \;|\; Z=z; \theta)]
$$

## Evidence

The term **evidence** in evidence lower bound is just a name for the likelihood function evaluated at a fixed $\theta$. 
$$
    \text{evidence} := \log p(X ; \theta)
$$


```python
def gaussian(x, mu, sigma):
    return 1/sigma*(2*torch.pi)**0.5 * torch.exp(-1/2*((x-mu)/sigma)**2)

x = torch.linspace(-5, 5, 1000)
fx = gaussian(x, mu=0, sigma=1)

plt.plot(x, fx, label='p(x)')
plt.plot(x, fx.log(), label='log p(x)')
plt.legend()
```
